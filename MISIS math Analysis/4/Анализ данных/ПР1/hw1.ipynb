### Уберём выбросы исп. м. межквартильного размаха IQR
Для определения выбросов потребуется:
1. Рассчитать квартили
2. Определить границы выбросов
3. Убрать сами выбросы

```python
# Одним из простых способов обнаружить выбросы является использование 
# метода межквартильного размаха IQR 

# Пример их рассчёта:
data = {'value': [10, 12, 12, 13, 12, 11, 45, 12, 11, 14, 13, 10]} 
test_df = pd.DataFrame(data) 

print("Исходные данные:") 
print(test_df.head(3)) 

# Вычисляем первый и третий квартиль
Q1 = test_df['value'].quantile(0.25)
Q3 = test_df['value'].quantile(0.75)
IQR = Q3 - Q1

# Определяем границы для выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Убираем выбросы
df_cleaned = test_df[(test_df['value'] >= lower_bound) & (test_df['value'] <= upper_bound)]
df_cleaned
```
